#### *DISPLAY*

```python
#Total Sales
total_sales=df['price'].sum()

#Average Sales
avg_sales=df['price'].mean()

#No.of items sold
no_of_items_sold=df['quantity'].count()

#Display
print(f"Total Sales: ${total_sales:,.3f}")
print(f"Average Sales: ${avg_sales:,.3f}")
print(f"No.of items sold: ${no_of_items_sold:,.3f}")
```

#### *VISUALIZATIONS*

```python
avg_price_by_category = df.groupby('category')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,6))
avg_price_by_category.plot(kind='bar', color='darkorange', edgecolor='black')
plt.title('Average Price by Category')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['sales'] = df['quantity'].astype(float) * df['price'].astype(float)
sales_by_gender = df.groupby('gender')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(6,5))
ax = sales_by_gender.plot(kind='bar', color='orchid', edgecolor='black')
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('')
plt.xticks(rotation=0)
plt.tick_params(axis='y', left=False, labelleft=False)
plt.tight_layout()

for i, val in enumerate(sales_by_gender):
    ax.text(i, val, f'{val:,.0f}', ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
df['sales'] = df['quantity'].astype(float) * df['price'].astype(float)
sales_by_payment = df.groupby('payment_method')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
ax = sales_by_payment.plot(kind='bar', color='goldenrod', edgecolor='black')
plt.title('Total Sales by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('')
plt.xticks(rotation=45)
plt.tick_params(axis='y', left=False, labelleft=False)
plt.tight_layout()

for i, val in enumerate(sales_by_payment):
    ax.text(i, val, f'{val:,.0f}', ha='center', va='bottom', fontsize=10)

plt.show()
```
